**Exercise 1**

```python
# Plotting: Exercise 1

import matplotlib.pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = 0.25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='darkred', linestyle='--',
             marker='o', markerfacecolor='yellow',
             ecolor='black', capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=0.25)
plt.legend(loc='lower right')
plt.show()
```

```python
# Plotting: Exercise 2

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Generate x data (range 0 to 50)
x = np.linspace(0, 50, 400)

# Generate y data
y = np.exp(-x / 10) * np.sin(x)

# Create the plot
plt.plot(x, y)

# Customize the plot
plt.xlabel('x', fontsize=15, color='darkorange')
plt.ylabel('y', fontsize=15, color='darkorange', rotation='horizontal')  # Corrected line
plt.title(r'$y = e^{-x/10} \cdot sin(x)$', fontsize=20)
plt.grid(alpha=0.25)
plt.ylim([-1, 1])

# Set y-axis ticks to increments of 0.25
plt.yticks(np.arange(-1, 1.01, 0.25))

# Format y-axis ticks to two decimal places
plt.gca().yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))

plt.show()
```
